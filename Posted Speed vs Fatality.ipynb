```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

raw_data= "Resources/table_02_29_040120.csv"
raw_data= pd.read_csv(raw_data)
raw_data
```

```python
#Get rid of unused rows

clean_data = raw_data.drop(raw_data.index[[0,1,14]])

clean_data
```

```python
# High vs. Low speed (Under 55mph) Posted Speed vs Fatality Totals

clean_data = clean_data.set_index('Category')

clean_data
```

```python
low_speed_deaths = [12040, 11859, 12419, 12077, 11742, 12571, 13417, 13673, 12812] 
high_speed_deaths = [14945, 14347, 14632, 14085, 14157, 15264, 16218, 15753, 15402]

low_speed_deaths_total = np.sum(low_speed_deaths)
high_speed_deaths_total = np.sum(high_speed_deaths)
difference = (high_speed_deaths_total - low_speed_deaths_total)

print(f"Sum of total deaths under 55mph : {low_speed_deaths_total}")
print(f"Sum of total deaths over  55mph : {high_speed_deaths_total}")
print(f"Difference in between low and high speed fatalities : {difference}")
```

```python
deaths = [12974, 34121, 51880, 11635, 76049, 9504, 24787, 17333, 7130]
index = ['5-25', '30-35', '40-45','50', '55', '60', '65', '70', '70+']
yearly_data = pd.DataFrame({'Deaths': deaths}, index=index)
ax = yearly_data.plot.bar(rot=0)
plt.title("Total Fatalities 2010-2018")
plt.xlabel('Posted Speed Limit')  
plt.ylabel('Fatalities')
plt.savefig('yearly_data.png')
```
